```python
#import frameworks
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("placement.csv")
```

```python
print(df.shape)
df.head()
```

```python
#plot with scatter plot
sns.scatterplot(x=df["cgpa"],y=df["iq"],hue=df["placement"])
```

```python
#creatinf x and y variables
#x=df.drop(columns=["Placement"])
# x=df.iloc[:,0:2]
#y=df["Placement"]
# y=df.iloc[:,-1]

X = df[['cgpa', 'iq']].values
y = df['placement'].values
```

```python
print(y)
```

```python
print(X)
```

```python
from sklearn.linear_model import Perceptron
P=Perceptron(max_iter=10000000, eta0=0.01, random_state=42)
```

```python
#Training a perceptron
P.fit(X,y)
```

```python
#Lets see weights of perceptron
print("Weights:", P.coef_)
print("Bias:", P.intercept_)
```

```python
#no lets plot decesion regions
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(8, 6))
plot_decision_regions(X, y, clf=P, legend=2)

plt.xlabel('CGPA')
plt.ylabel('IQ Score')
plt.title('Perceptron Decision Boundary for Placement Prediction')
plt.show()
```

```python
#Perceptron Equation
w = P.coef_[0]     # [w1, w2]
b = P.intercept_[0] # bias
print( "Bias:",b,"+ W1:",w[0],"x X1 + ", "W2:",w[1],"x X2")
```
